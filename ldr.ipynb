```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
receitas = [10000, 12000, 11000, 10500, 13000, 14000, 15000, 16000, 15500, 17000, 17500, 18000]
despesas = [8000, 8500, 9000, 9500, 10000, 10500, 11000, 11500, 12000, 12500, 13000, 13500]

# Calcular os lucros
lucros = [receita - despesa for receita, despesa in zip(receitas, despesas)]

# Criar um DataFrame com os dados
df = pd.DataFrame({'Meses': meses, 'Receitas': receitas, 'Despesas': despesas, 'Lucros': lucros})

# Definir a coluna 'Meses' como índice
df.set_index('Meses', inplace=True)

# Plotar a série temporal
plt.figure(figsize=(10, 6))
df.plot(kind='line', marker='o')
plt.title('Série Temporal de Receitas, Despesas e Lucros')
plt.xlabel('Meses')
plt.ylabel('Valor')
plt.xticks(rotation=45)
plt.grid(True)

# Adicionar linhas verticais para sinalizar o início do primeiro quadrimestre e do terceiro quadrimestre
plt.axvline(x=3.5, color='red', linestyle='--', linewidth=1)  # Linha para marcar o início do primeiro quadrimestre (Abril)
plt.axvline(x=7.5, color='blue', linestyle='--', linewidth=1)  # Linha para marcar o início do terceiro quadrimestre (Agosto)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Dados
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
receitas = [10000, 12000, 11000, 10500, 13000, 14000, 15000, 16000, 15500, 17000, 17500, 18000]

# Gerar valores aleatórios para as despesas de diferentes categorias
np.random.seed(0)  # Define a semente aleatória para reproducibilidade
despesas_pessoal = np.random.randint(6000, 8000, 12)  # Despesas com Pessoal
despesas_correntes = np.random.randint(4000, 6000, 12)  # Despesas Correntes
despesas_capital = np.random.randint(2000, 4000, 12)  # Despesas de Capital
despesas_transferencias = np.random.randint(1000, 3000, 12)  # Despesas de Transferências

# Calcular os lucros
lucros = [receita - sum(despesas) for receita, despesas in zip(receitas, 
                                                              zip(despesas_pessoal, 
                                                                  despesas_correntes, 
                                                                  despesas_capital, 
                                                                  despesas_transferencias))]

# Criar um DataFrame com os dados
df = pd.DataFrame({'Meses': meses, 
                   'Receitas': receitas, 
                   'Despesas Pessoal': despesas_pessoal, 
                   'Despesas Correntes': despesas_correntes, 
                   'Despesas de Capital': despesas_capital, 
                   'Despesas de Transferências': despesas_transferencias,
                   'Lucros': lucros})

# Definir a coluna 'Meses' como índice
df.set_index('Meses', inplace=True)

# Plotar a série temporal
plt.figure(figsize=(12, 8))
df[['Receitas', 'Despesas Pessoal', 'Despesas Correntes', 'Despesas de Capital', 'Despesas de Transferências']].plot(kind='line', marker='o', ax=plt.gca())
plt.title('Série Temporal de Receitas e Despesas por Categoria')
plt.xlabel('Meses')
plt.ylabel('Valor')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Dados
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

# Gerar valores aleatórios para cada tipo de receita
np.random.seed(0)  # Define a semente aleatória para reproducibilidade
receitas_tributarias = np.random.randint(10000, 20000, 12)
receitas_transferencias = np.random.randint(5000, 15000, 12)
receitas_patrimoniais = np.random.randint(1000, 5000, 12)
receitas_servicos = np.random.randint(2000, 8000, 12)
receitas_creditos = np.random.randint(2000, 10000, 12)
outras_receitas = np.random.randint(1000, 7000, 12)

# Calcular a soma total das receitas
receitas_totais = [sum(x) for x in zip(receitas_tributarias, receitas_transferencias, receitas_patrimoniais, receitas_servicos, receitas_creditos, outras_receitas)]

# Criar um DataFrame com os dados
df = pd.DataFrame({'Meses': meses,
                   'Receitas Tributárias': receitas_tributarias,
                   'Receitas de Transferências': receitas_transferencias,
                   'Receitas Patrimoniais': receitas_patrimoniais,
                   'Receitas de Serviços': receitas_servicos,
                   'Receitas de Créditos': receitas_creditos,
                   'Outras Receitas': outras_receitas,
                   'Receitas Totais': receitas_totais})

# Definir a coluna 'Meses' como índice
df.set_index('Meses', inplace=True)

# Plotar a série temporal
plt.figure(figsize=(12, 8))
df[['Receitas Tributárias', 'Receitas de Transferências', 'Receitas Patrimoniais', 'Receitas de Serviços', 'Receitas de Créditos', 'Outras Receitas', 'Receitas Totais']].plot(kind='line', marker='o', ax=plt.gca())
plt.title('Série Temporal de Receitas por Categoria')
plt.xlabel('Meses')
plt.ylabel('Valor')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Dados
valor_emprestimo = 240000
parcelas = 12
taxa_juros_nominal = 0.12  # 12% ao ano

# Calculando a taxa de juros mensal
taxa_juros_mensal = taxa_juros_nominal / 12

# Calculando a parcela mensal (método Price)
parcela_price = valor_emprestimo * (taxa_juros_mensal / (1 - (1 + taxa_juros_mensal)**-parcelas))

# Método de Amortização Misto
# Calculando a amortização constante
amortizacao_constante = valor_emprestimo / parcelas

# Calculando as parcelas de juros para o método misto
juros_misto = []
saldo_devedor_misto = [valor_emprestimo]
for i in range(parcelas):
    juros = saldo_devedor_misto[-1] * taxa_juros_mensal
    juros_misto.append(juros)
    saldo_devedor = saldo_devedor_misto[-1] - amortizacao_constante + juros
    saldo_devedor_misto.append(saldo_devedor)

# Criando as séries temporais
serie_price = pd.Series([parcela_price] * parcelas, index=range(1, parcelas + 1))
serie_misto = pd.Series([amortizacao_constante + juros for juros in juros_misto], index=range(1, parcelas + 1))
serie_saldo_misto = pd.Series(saldo_devedor_misto[:-1], index=range(1, parcelas + 1))

# Exibindo as séries temporais
print("Série Temporal - Método Price:")
print(serie_price)
print("\nSérie Temporal - Método Misto (Parcela):")
print(serie_misto)
print("\nSérie Temporal - Método Misto (Saldo Devedor):")
print(serie_saldo_misto)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
marcos_orcamentarios = ['PPA', 'LDO', 'LOA']  # Plano Plurianual, Lei de Diretrizes Orçamentárias, Lei Orçamentária Anual

# Gerar uma série com os marcos do ciclo orçamentário
marcos_orcamentarios_series = pd.Series([''] * len(meses), index=meses)

# Definir os meses em que cada marco ocorre
marcos_orcamentarios_series['Janeiro'] = 'PPA'
marcos_orcamentarios_series['Fevereiro'] = 'LDO'
marcos_orcamentarios_series['Março'] = 'LOA'

# Criar um DataFrame com os dados
df = pd.DataFrame({'Meses': meses, 'Marcos Orçamentários': marcos_orcamentarios_series})

# Definir a coluna 'Meses' como índice
df.set_index('Meses', inplace=True)

# Plotar a série temporal
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Marcos Orçamentários'], marker='o', linestyle='', color='blue')
plt.title('Marcos do Ciclo Orçamentário')
plt.xlabel('Meses')
plt.ylabel('Marcos Orçamentários')
plt.yticks([])  # Remover os valores do eixo y
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
